## 7) How much money do these top 5 billionaires have in total?

> **TIP:** for multiple conditions, you can't use `or` and `and`. In pandas, `|` means *or* and `&` means *and*.
>
> **TIP:** Checking multiple conditions can be done two ways.
> - `(df.colname == 'A') | (df.colname == 'B')`
> - `df.colname.isin(['A', 'B'])`

```python
# This line of code gives me the top 5 billionaires.

df.sort_values(by='networthusbillion', ascending=False).head(5)
```

```python
# This line gives me the total money ALL the billionaires have.

df['networthusbillion'].sum()
```

```python
# So the answer to the question would be combining them in some way.

# df[(df.sort_values(by='networthusbillion', ascending=False).head(5)) & (df['networthusbillion'].sum())] I tried to overcomplicate again.

df.sort_values(by='networthusbillion', ascending=False).head(5)['networthusbillion'].sum()
```
